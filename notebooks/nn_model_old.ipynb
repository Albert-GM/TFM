```python
from keras.models import Sequential
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/Users/agm/Documents/kschool/TFM/data_out/data_to_nn.csv")
df.head()
```

```python
X = df.drop('total_death', axis = 1)
y = df['total_death']
```

```python
display(X)
display(y)
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(), ['country_code']), remainder='passthrough')
X = column_trans.fit_transform(X)
```

```python
model = Sequential()
```

```python
from keras.layers import Dense

model.add(Dense(units=64, activation='relu', input_dim=63))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='softmax'))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='sgd',
              metrics=[['mae', 'acc']])
```

```python
X.shape
```

```python
y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y.values, random_state=0) 
```

```python
# x_train and y_train are Numpy arrays --just like in the Scikit-Learn API.
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
loss_and_metrics = model.evaluate(X_train, y_train, batch_size=128)
```

```python
loss_and_metrics
```

```python
loss_and_metrics = model.evaluate(X_test, y_test, batch_size=128)
```
